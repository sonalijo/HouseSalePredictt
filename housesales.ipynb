```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
labels = data['price']
conv_dates = [1 if values == 2014 else 0 for values in data.date ]
data['date'] = conv_dates
train1 = data.drop(['id', 'price'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.10,random_state =2)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
from sklearn import ensemble

clf = ensemble.GradientBoostingRegressor(
    n_estimators=400,
    min_samples_split=2,
    learning_rate=0.1,
    loss='squared_error'   
)
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test,y_test)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

t_sc = np.zeros(clf.n_estimators, dtype=np.float64)

for i, y_pred in enumerate(clf.staged_predict(x_test)):
    # Har stage pe MSE calculate karo
    t_sc[i] = mean_squared_error(y_test, y_pred)

print("Minimum test MSE:", t_sc.min())
```

```python
y_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# test error calculation
test_score = []
for y_pred in clf.staged_predict(x_test):
    test_score.append(mean_squared_error(y_test, y_pred))

# number of estimators
testsc = np.arange(clf.n_estimators_) + 1

# plot
plt.figure(figsize=(12,6))
plt.plot(testsc, test_score, label='Test Error')
plt.xlabel("n_estimators")
plt.ylabel("MSE")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(testsc,clf.train_score_,'b-',label= 'Set dev train')
plt.plot(testsc,t_sc,'r-',label = 'set dev test')
```
